### jieba分词 + 停用词典法

```python
import jieba 
# 待分词的中文文本
text = "自然语言处理是人工智能领域中的一个重要方向，它研究能实现人与计算机之间用自然语言进行有效通信的各种理论和方法。"

# 使用精确模式进行分词
words_precise = jieba.cut(text, cut_all=False)
print("精确模式分词结果: ", list(words_precise))

# 使用搜索引擎模式进行分词
words_search = jieba.cut_for_search(text)
print("搜索引擎模式分词结果: ", list(words_search))

# 使用全模式进行分词，会列出所有可能的词语
words_all = jieba.cut(text, cut_all=True)
print("全模式: ", list(words_all))

#精确模式：切分出的词语更符合语言习惯，通常用于文本分析。
#搜索引擎模式：在精确模式的基础上，对长词再次切分，提高召回率，适合搜索引擎索引构建。
#全模式：会将所有可能的词语切分出来，包括单字，常用于文本的全面分析。
```

```python
# 创建一个空列表，用于存储从停用词表文件中读取的停用词
stop_list = []

# 使用 with 语句打开停用词表文件，确保文件会在操作完成后自动关闭
# 'r' 表示以只读方式打开文件，encoding='utf-8' 表示使用 utf-8 编码读取文件
with open('停用词表.txt','r', encoding = 'utf-8') as f:
    # 读取文件的所有行
    for line in f.readlines():
        # 移除每行末尾的换行符 '\n'，并将处理后的词语追加到 stop_list 列表中
        stop_list.append(line.replace('\n',''))

print(stop_list)
```

```python
# 打开一个名为 'cut.csv' 的文件用于写入，如果文件不存在则创建它
# 'w' 表示以写入模式打开文件，如果文件存在则覆盖，encoding='utf-8' 表示使用 utf-8 编码写入文件
cut = open('cut.csv','w', encoding = 'utf-8')

# 遍历 DataFrame 'df' 中的每一行，对于每一行，idx 变量会获取当前行的索引，row 变量会获取当前行的数据，作为一个 Series 对象。
for idx, row in df.iterrows():
    # 使用 jieba.cut 方法对 'judge' 列中的字符串进行分词处理
    words = jieba.cut(str(row['judge']))

    # 遍历分词后生成的词语列表
    for word in words:
        # 如果当前词语不在停用词列表中，则将其写入 'cut.csv' 文件
        if word not in stop_list:
            cut.write(word + '')
        # 如果当前词语在停用词列表中，则打印出来（这里可能是用来调试，实际应用中可能不需要打印）
        else:
            print(word)

    # 每个 'judge' 列处理完成后，写入一个换行符，以便在 CSV 文件中分隔不同的行
    cut.write('\n')

# 关闭 'cut.csv' 文件，释放资源
cut.close()
```

```python
# 确保停用词列表不包含NaN，并转换为集合以提高搜索效率
stop_list = {word for word in stop_list if pd.notna(word)}

# 打开文件准备写入，使用'w'模式打开文件，使用utf-8-sig编码
with open('cut.csv', 'w', encoding='utf-8-sig') as cut:
    # 遍历DataFrame 'df'中的每一行
    for idx, row in df.iterrows():
        # 检查'judge'列是否为NaN，如果是NaN，则跳过当前行
        if pd.isna(row['judge']):
            continue
        
        # 使用jieba.cut方法对'judge'列中的字符串进行分词处理
        words = jieba.cut(str(row['judge']))
        
        # 将分词后的结果过滤掉停用词，并用空格连接成字符串
        filtered_words = ' '.join(word for word in words if word not in stop_list)
        
        # 写入过滤后的词语到文件，每个样本后跟一个换行符
        cut.write(filtered_words + '\n')

# 注意：这里不再需要显式调用cut.close()，因为使用了with语句，文件会在with块结束时自动关闭
```

```python
#######分词方案二：
# 初始化一个停用词列表，先添加一个空字符串作为占位符
stop_list = ['']

# 打开名为 '停用词表.txt' 的文件，以只读模式读取，并指定文件编码为 utf-8
with open('停用词表.txt', 'r', encoding='utf-8') as f:
    # 读取文件的所有行
    for line in f.readlines():
        # 移除每行末尾的换行符 '\n'，并将处理后的词语追加到 stop_list 列表中
        stop_list.append(line.replace('\n', ''))

# 初始化一个空列表，用于存储处理后的分词字符串
data_list = []

# 遍历 DataFrame 'df' 中的 'judge' 列
for i in df['judge']:
    # 使用 jieba.cut 方法对 'judge' 列中的每个元素进行分词处理
    # 并创建一个新的列表，其中包含不在停用词列表中的词语
    # 如果分词结果在停用词列表中，则用空字符串 '' 替换
    data_list_temporary = [x if x not in stop_list else '' for x in jieba.cut(str(i))]
    
    # 移除列表中所有的空字符串 ''
    while '' in data_list_temporary:
        data_list_temporary.remove('')
    
    # 将处理后的词语列表转换为一个字符串
    # 使用 str() 将列表转换为字符串，并用 [1:-1] 去掉首尾的方括号
    # 然后替换所有的单引号 '，因为列表转换为字符串时，每个元素会被单引号包围
    data_list_temporary_str = str(data_list_temporary)[1:-1].replace("'", '').replace("'", '')
    
    # 将处理后的词语字符串添加到 data_list 中
    data_list.append(data_list_temporary_str)

# 打印 data_list，这个列表包含了处理后的所有词语字符串
print(data_list)

# 将 data_list 中的数据赋值给 DataFrame 'df' 的新列 'judge_cut'
df['judge_cut'] = data_list

# 将包含新列 'judge_cut' 的 DataFrame 保存到 CSV 文件中
# 指定了完整的文件路径，文件名，不保存索引，以及使用 utf-8-sig 编码
df.to_csv(r'C:\Users\12023\Desktop\课程文件\计算社会学\明星信息1.csv', index=False, encoding='utf-8-sig')
```

# 主题模型分析

隐含狄利克雷分布(Latent Dirichlet Allocation，LDA)，是一种主题模型(topic model)。LDA是一种非监督学习技术，可以用来识别大规模文档集(documentds collection)或语料库(corpus)中潜藏的主题信息。它采用了词袋(bag of words)的方法,这种方法将每一篇文档视为一个词频向量，从而将文本信息转化为了易于建模的数字信息。但是词袋方法没有考虑词与词之间的顺序，这简化了问题的复杂性，同时也为模型的改进提供了契机。每一篇文档代表了一些主题所构成的一个概率分布，而每一个主题又代表了很多单词所构成的一个概率分布。

LDA是-个Bayes Hierarchy Model。所谓Bayes Hierarchy Model说白了就是把莫型的参数看作随机变量，这样可以引入控制参数的参数。主题的一个通式为
P(w|d) = sigma{ p(w|z)*p(z|d) }

其中主题topic，说白了就是一个一元语言模型，没有任何特殊的地方。对应上面的公式，就是p(wz)。而topic model，一般指的是两种分布:第一种就是topic~word的分布，就是p(w|z)，表示word出现在topic中的概率。

第二种是p(z|d)，这个是doc~topic分布，表示topic出现在doc中概率。有了这两种分布后，这个文档集合就有了一种立体化的感觉：



一个三层的文档表示空间跃然纸上。而最上层，就是人们常说的“降维”，其实是把文档投影到了“topic”空间。doc-topic-word这个Bayes Chain，就可以涵盖LDA的一个最基本的思想。
对于语料库中的每篇文档，LDA定义了如下生成过程(generative process）：
1.对每一篇文档，从主题分布中抽取一个主题:
2.从上述被抽到的主题所对应的单词分布中抽取一个单词，
3.重复上述过程直至遍历文档中的每一个单词。
